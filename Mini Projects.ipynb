```python
# Guess the Number(user)
import random
def random_number(x):
    random_number = random.randint(1,x)
    guess_number = 0
    while guess_number!=random_number:
        guess_number = int(input(f'Guess a number between 1 and {x}: '))
        if guess_number > random_number:
            print('Too high. Guess Again')
        elif guess_number < random_number:
            print('Too Low. Guess Again')
    
    print('Yay! You guessed it correct')

random_number(20)
```

```python
# Guess the Number(Computer)
import random

def computer_number(x):
    low=0
    high=x
    feedback = ' '
    while feedback!='c':
        if low!=high:
            computer_guess = random.randint(low, high)
        else:
            computer_guess = low
        feedback = input(f'is {computer_guess} too high(h) or too low(l) or correct(c)?: ')
        if feedback=='h':
            high = computer_guess-1
            computer_guess = random.randint(low, high)
        elif feedback=='l':
            low = computer_guess+1
            computer_guess = random.randint(low, high)
    
    print('Yay! the computer guessed it correct')
    
computer_number(200)
```

```python
# Word Guessing Game
import random
#list of words
words = ["lemon","mango","banana","apple",'kiwi','watermellon','strawberry','guava','cherry','orange','pineapple']
#ask Name
name=input("Enter  Your name: ")
print(f"Hi {name} , Let's start the game")

print("\n\n\n\t\t ***Word Guessing Game*** ")
print("\n You have 10 attempts to guess a fruit correctly .")
#list to store wrong guessed_letters
wrong_list=[]
#generate random word
original_word=random.choice(words)
print(f"The word is of {len(original_word)} letters. ")
#create an empty list
guessed_word = []
for i in range(len(original_word)):
    guessed_word.append("__")
print(*([i for i in guessed_word]))
c=8
while(c):
    c=c-1 
    #take input from user
    guessed_letter = input("Guess the  letter\n")
    #check if guessed_letter is an alphabet
    if not guessed_letter.isalpha():
        print('Guess only a letter')
    #check if guessed letter length is one or not
    elif(len(guessed_letter)>1):
        print('Guess only one letter...')
    #check that letter chosen by user is already guessed or not
    elif(guessed_letter in wrong_list):
        print('You have Already guessed this letter')
    #check if guessed_letter is matches with original_word
    if guessed_letter in original_word:
        for i in range(len(original_word)):
            if original_word[i] == guessed_letter:
                guessed_word[i]=original_word[i] 
    else:
        """if guessed_letter is not in original_word 
           prompt user for wrong chosen letter"""
        print("You Guessed wrong letter")
        wrong_list.append(guessed_letter)
    guess_word = [i for i in guessed_word]
    guess_word = "".join(guess_word)
    if original_word ==guess_word :
        print('YAY !!, You have Got the letter ...')
        print()
        print(f'The original Word was {(original_word).upper()}')
        break
    #print the guessed word
    print(*([i for i in guessed_word]))
    #prompt user showing number of attempts left to win
    print(f"You have {c} attempts left")
    if c==0:
        if  original_word!=guessed_word  :
            print(f"You lost the game ,The original Word was {original_word}")
            exit(0)
```

```python
# Binary Search Alogrithm
def binary(arr, x):
    low = 0
    high = len(arr)-1
    
    while low <= high:
        mid = (low+high)//2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1

result = binary([10,20,30,40,50,60], 30)

if result !=-1:
    print('Element present in array')
else:
    print('Element not present')
```

```python
# leap year checker
def leap_year():
    year = int(input('Enter Year: '))

    # To get year (integer input) from the user
    # year = int(input("Enter a year: "))

    if (year % 4) == 0:
        if (year % 100) == 0:
            if (year % 400) == 0:
                print("{0} is a leap year".format(year))
            else:
                print("{0} is not a leap year".format(year))
        else:
            print("{0} is a leap year".format(year))
    else:
        print("{0} is not a leap year".format(year))

        
leap_year()
```
